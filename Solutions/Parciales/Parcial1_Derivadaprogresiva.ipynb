```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.01
x = np.arange(0.1,1.1,h)
```

```python
func = lambda x: np.sqrt(np.tan(x))
derivada = lambda x: (  1  ) / ( 2 * np.cos(x)**2 * np.sqrt( np.tan(x) ) ) 
```

```python
plt.plot(x,derivada(x))
```

```python
def CentralDerivative(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)
```

# Derivada progresiva de orden $\mathcal{O}(h^{2})$

$$ f'(x) \simeq \frac{1}{2h}( -3f(x) + 4f(x+h) - f(x+2h)) $$

```python
def ProgresiveDerivative(x,f,h):
    return (-3*f(x)+4*f(x+h)-f(x+2*h))/(2*h)
```

```python
plt.plot(x,derivada(x),label='Exacta')
plt.plot(x,CentralDerivative(x,func,h),label='Central')
plt.plot(x,ProgresiveDerivative(x,func,h),label='Second order progresive')
plt.legend(loc=0)
```

```python
plt.plot(x,np.abs(derivada(x)-CentralDerivative(x,func,h)),label='Central')
plt.plot(x,np.abs(derivada(x)-ProgresiveDerivative(x,func,h)),label='Second order progresive')
plt.legend(loc=0)
```
