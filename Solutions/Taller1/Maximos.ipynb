```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('EstrellaEspectro.txt')
```

```python
x = data[:,0]
y = data[:,1]
```

```python
plt.plot(x,y)
```

```python
maxima = []
imagenes = []

for i in range(1,len(x)-1):
    if y[i] > y[i-1] and y[i] > y[i+1]:
        maxima.append(x[i])
        imagenes.append(y[i])
```

```python
plt.plot(x,y)

for i in range(len(maxima)):
    plt.scatter(maxima,imagenes,color='r',marker='.')
    #plt.axvline(x = maxima[i],ls='--',color='r')
    
#plt.savefig('Maximos.pdf')
```
