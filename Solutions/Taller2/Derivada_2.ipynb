```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h= 0.05
N = int(20/h)+1
print(N)
```

```python
x = np.linspace(-10,10,N)
function = lambda x: 1/np.sqrt( 1 + np.exp(-x**2) )
```

```python
def GetDerivative(x,f,h):
    return (f(x+h)-f(x-h))/(2.*h)
```

```python
Derivative = GetDerivative(x,function,h)
```

```python
plt.plot(x,Derivative)
```

```python
ExactDf = lambda x: (x*np.exp(-x**2))/( 1 + np.exp(-x**2) )**(3/2)
ExactDerivative = ExactDf(x)
```

```python
plt.plot(x,Derivative,color='r',label='Right Derivative')
plt.plot(x,ExactDerivative,color='b', label='Exact Derivative')
plt.legend()
```

# Error

```python
plt.plot(x,np.abs(Derivative-ExactDerivative))
```
