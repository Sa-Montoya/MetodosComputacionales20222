```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h= 0.05
N = int(20/h)+1
print(N)
```

```python
x = np.linspace(-10,10,N)
function = lambda x: 1/np.sqrt( 1 + np.exp(-x**2) )
```

```python
def GetDerivative(x,f,h):
    return (f(x+h)-f(x-h))/(2.*h)
```

```python
Derivative = GetDerivative(x,function,h)
```

```python
plt.plot(x,Derivative)
```

# Convolucion, encontramos las imagenes

```python
f = function(x)
#print(f)
```

```python
M = np.array([+1,0,-1])
```

```python
def GetConvolutionDerivative(f,M,h):
    
    Derivative = np.zeros_like(f)
    
    for i in range(1,len(f)-1):
        
        deriva = 0.
        
        for j in range(-1,2):
            #print(i,j)
            deriva += M[j+1]*f[i-j]
            
        Derivative[i] = deriva/(2*h)
        
    return Derivative
```

```python
ConvolDerivative = GetConvolutionDerivative(f,M,h)
```

```python
plt.plot(x,Derivative,color='r',label='Right Derivative')
plt.plot(x,ConvolDerivative,color='b',label='Convolve Derivative')
```
