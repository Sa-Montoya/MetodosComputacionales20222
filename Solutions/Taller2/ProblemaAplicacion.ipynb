```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,25)
y = x.copy()
h = x[1] - x[0]
print(h)

R = 2
```

```python
def Potencial(x,y,V=2,R=2):
    return V*x*(1 - R**2/(x**2+y**2))
```

```python
def Dx(f,x,y,h):
    return (f(x+h,y) - f(x-h,y))/(2*h)
def Dy(f,x,y,h):
    return (f(x,y+h) - f(x,y-h))/(2*h)
```

```python
Vx = np.zeros((len(x),len(y)))
Vy = np.zeros((len(x),len(y)))
```

```python
for i in range(len(x)):
    for j in range(len(x)):
        
        if (x[i]**2 + y[j]**2 > R**2):
        
            Vx[i,j] = Dx(Potencial,x[i],y[j],0.001)
            Vy[i,j] = -Dy(Potencial,x[i],y[j],0.001)
```

```python
fig1 = plt.figure(figsize=(10,10))
ax = fig1.add_subplot(1,1,1)

circle = plt.Circle((0,0),R,color='r',lw=2,fill=False)

ax.add_artist(circle)

for i in range(x.shape[0]):
    for j in range(y.shape[0]):
        if (x[i]**2 + y[j]**2 > R**2):
            ax.quiver(x[i],y[j],Vx[i,j],Vy[i,j],color='b',alpha=0.7)
        
        
ax.set_xlabel(r'$x[cm]$',fontsize=15)
ax.set_ylabel(r'$y[cm]$',fontsize=15)

plt.savefig('FlujoCilindro.pdf')
```
