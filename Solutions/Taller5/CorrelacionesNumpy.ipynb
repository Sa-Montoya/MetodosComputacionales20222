```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from tqdm import tqdm
```

```python
def Ckfunction(vector, Npoints_, moment):
    
    ck = []
    
    for k in tqdm(range(moment)):
        k += 1
        sum_ = 0.
        
        for i in range(len(vector[:-k])):
            sum_ += vector[i]*vector[i+k]
            
        ck.append(sum_/(Npoints_-1))

    return ck
```

```python
U = np.random.rand(10000)

Cksimple = Ckfunction(U,10000, 30)
```

```python
fig = plt.figure(figsize = (10,10))

ax1 = fig.add_subplot(1,1,1)

ax1.plot(Cksimple, linewidth=3)

ax1.set_title('Generador Numpy')
ax1.set_xlabel('k-esimo vecino', fontsize=12)
ax1.set_ylabel('C(k)', fontsize=15)

ax1.set_ylim(0.20,0.30)

ax1.grid()

plt.savefig('CkNumpy.pdf')
```
