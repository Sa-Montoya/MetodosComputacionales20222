```python
def E(x):
    #return np.mean(x)
    return np.sum(x)/(len(x)-1)

def GetCov(data):
    
    N = data.shape[1]
    
    Cov = np.zeros((N,N))
    
    for i in range(N):
        for j in range(N):
            
            X = data[:,i]
            Y = data[:,j]
            
            Cov[i,j] = E( (X-E(X))*(Y-E(Y)) )
            
    return Cov
```
