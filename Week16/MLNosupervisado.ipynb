```python
import numpy as np
import matplotlib.pyplot as plt

import mglearn
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

import pandas as pd
import seaborn as sns
```

```python
X,Y = make_blobs(n_features=10, n_samples = 100, centers=5, random_state=1, cluster_std=1 )
Y.shape
```

```python
df = pd.DataFrame(X, index=range(X.shape[0]),columns=range(X.shape[1]))
#df
```

```python
df['Cluster']=Y
#df
```

```python
Dict = { 0:'X1', 1:'X2', 2:'X3', 3:'X4', 4:'X5', 5:'X6', 6:'X7', 7:'X8', 8:'X9', 9:'X10'}
```

```python
df.rename(columns=Dict, inplace=True)
#df
```

```python
sns.lmplot( x = Dict[0],y = Dict[2], data=df, hue='Cluster',fit_reg=False )
```

```python
N = 4
pca = PCA(n_components=N)
pc = pca.fit_transform(X)
```

```python
df1 = pd.DataFrame({'Var':pca.explained_variance_ratio_,'PC':['PC1','PC2','PC3','PC4']})
```

```python
sns.barplot(x='PC',y='Var',data=df1)
```

```python
pc_df = pd.DataFrame(data=pc, columns=['PC1','PC2','PC3','PC4'])
pc_df['Cluster'] = Y
```

```python
sns.lmplot( x = 'PC1',y = 'PC3', data=pc_df, hue='Cluster',fit_reg=False )
```
